# FastAPI — Complete Guide


## 1. What is FastAPI?

**FastAPI** is a modern, high-performance Python **web framework** used for building APIs quickly and efficiently. It was created by **Sebastián Ramírez** and released in 2018.

FastAPI is built on top of two powerful libraries:
- **Starlette** — handles web (HTTP) routing and requests
- **Pydantic** — handles data validation and serialization

It uses Python's **type hints** to automatically validate data, generate documentation, and reduce bugs — making development faster and cleaner

## 2. Library vs Framework — What's the Difference?

This is one of the most common questions in software development. Here's the core distinction:

###  Library
A **library** is a collection of tools (functions, classes) that **you call** whenever you need them. You are in control of the flow of your program.

**Example:** The `requests` library in Python.
```python
import requests
response = requests.get("https://api.example.com/data")
```
You decide *when* and *where* to call it.

###  Framework
A **framework** provides a structure and calls **your code** at the right time. This is called **Inversion of Control (IoC)** — the framework is in charge, not you.

# 3. Why Do We Need FastAPI?

Before FastAPI, developers used frameworks like **Flask** or **Django REST Framework**. While these are great, they have some limitations that FastAPI solves:

### ⚡ Speed
FastAPI is one of the **fastest Python frameworks** available — on par with NodeJS and Go, thanks to **async** support via Python's `asyncio`.

### ✅ Automatic Data Validation
Using Python type hints, FastAPI automatically validates incoming data. No more writing manual validation code.
