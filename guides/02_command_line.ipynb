## Part 1: What is the Command Line?
The command line is a text-based interface that allows you to interact with your computer by typing commands, rather than using a mouse or graphical user interface (GUI).

### Tools on Windows
1. CMD (Command Prompt)
The traditional Windows command line.
Older, limited features.
Syntax differs from Unix (uses dir instead of ls, etc.)

2. PowerShell
Modern replacement for CMD.
More powerful scripting capabilities.
Supports most Unix-style commands (e.g., ls, cat, cd) via aliases.
This is what we’ll use during setup (e.g., installing uv environments).

3. Terminal in Cursor
Cursor (based on VSCode) includes an integrated terminal.
After setup, you’ll use this Terminal for daily tasks like running notebooks, scripts, and Git.
Can be configured to use PowerShell or another shell.

| Task              | Command          |
| ----------------- | ---------------- |
| List directory    | `ls`             |
| Print current dir | `pwd`            |
| Change dir        | `cd folder`      |
| Go up             | `cd ..`          |
| Show hidden files | `ls -a`          |
| Create dir        | `mkdir`          |
| Delete file       | `rm filename`    |
| Delete folder     | `rm -r folder`   |
| Copy              | `cp source dest` |
| Move/rename       | `mv old new`     |
| Read file         | `cat filename`   |
