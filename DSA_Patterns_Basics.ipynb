- What is O(n)? — Big O Notation
- Big O is simply a way to describe how fast or slow your code is as the input grows larger. It answers the question:

- "If I give this code more data, how much longer will it take?"

- Think of it like this real-life example
Imagine you're looking for a friend's name in a list.
Scenario 1 — 10 names in the list
You check each name one by one. Takes maybe 5 seconds.
Scenario 2 — 1000 names in the list
Now it takes much longer because there's more to check.
The relationship between the size of the list and the time it takes — that's what Big O measures.

O(1) — Constant Time
- No matter how big the input is, it always takes the same time.

```python
arr = [1, 2, 3, 4, 5]
print(arr[0])  # Always instant, doesn't matter how big arr is
```

O(n) — Linear Time
- As the input grows, the time grows at the same rate.

```python
arr = [1, 2, 3, 4, 5]
for num in arr:
    print(num)
```

O(n²) — Quadratic Time
- As the input grows, the time grows much faster (nested loops).

```python
arr = [1, 2, 3, 4, 5]
for i in arr:
    for j in arr:    # Loop inside a loop
        print(i, j)
```

1. Two Pointers
Real world: You have a sorted line of people by height. You want to find two people whose heights add up to exactly 6 feet. Instead of comparing everyone with everyone, one person starts from the shortest end, another from the tallest end, and they walk toward each other.

Sum too big? Tall person steps inward
Sum too small? Short person steps inward
Much faster than checking every pair!


2. Sliding Window
Real world: You're watching a CCTV recording and want to find the 3-minute clip with the most movement. Instead of rewatching the whole video for every possible 3-minute window, you slide a fixed window forward — add the new second, remove the old second, keep moving.
The "window" slides, never restarts from scratch.

3. Frequency Count
Real world: A teacher has a box of colored marbles and wants to know which color appears most. She goes through each marble and keeps a tally on paper. That tally sheet is your frequency count.

4. String Parsing
Real world: Reading a sentence word by word. Like when you read "My name is John" and your brain automatically splits it at spaces and processes each word. That's string parsing — going through characters and extracting meaningful pieces.

🟡 INTERMEDIATE PATTERNS

5. HashMap
Real world: A phone contacts app. You search "John" and instantly get his number. You don't scroll through every contact — the app maps Name → Number directly. That instant lookup is exactly what a HashMap does.
"John" → 9876543210
"Alice" → 1234567890

6. HashSet
Real world: A nightclub bouncer with a list of banned people. When someone arrives, he checks if their name is on the banned list — just yes or no. No duplicates, just unique names, instant check. That's a HashSet.

7. Prefix Sum
Real world: Your electricity bill shows monthly usage. At the end of the year you want to know total usage from March to August. Instead of adding those 6 months again, if you already have running totals saved (Jan total, Jan+Feb total, etc.), you just subtract two numbers.
Total[Aug] - Total[Feb] = Usage from March to August
Pre-calculate once, answer any range question instantly.

8. Prefix Sum + HashMap
Real world: You're tracking your daily spending and want to find which days your total spending hit exactly ₹5000 from the start. You combine the running total idea (prefix sum) with instant lookup (HashMap) to find it without checking every combination.

9. Binary Search
Real world: Finding a word in a physical dictionary. You don't start from page 1. You open the middle — if your word comes before, ignore the right half. Open the middle of the left half — keep halving until you find it. 1000 pages becomes just 10 steps.

10. Expand Around Center
Real world: You drop a stone in a pond. Ripples expand outward equally in all directions from the center. In palindrome problems, you pick a center character and expand left and right equally as long as both sides match.

11. Index Marking
Real world: A teacher marks attendance directly on the student list using checkmarks. Instead of maintaining a separate sheet, she uses the original list itself to store information. Similarly, index marking uses the array itself to track what's been seen.

🔴 ADVANCED PATTERNS

12. Kadane's Algorithm
Real world: You're a street food vendor tracking daily profit/loss over a month. You want to find the best continuous stretch of days where you made the most profit overall. Kadane's helps you find that best streak efficiently — keep adding days as long as it helps, restart when it hurts.
Days:   Mon  Tue  Wed  Thu  Fri
Profit: +50  -20  +80  -10  +40

Best streak: Wed to Fri = +110

13. DP (Dynamic Programming)
Real world: You're climbing a staircase and can take 1 or 2 steps at a time. How many ways to reach step 10? Instead of recalculating from scratch every time, you remember:

Ways to reach step 1 = 1
Ways to reach step 2 = 2
Ways to reach step 3 = step1 + step2 = 3
Ways to reach step 4 = step2 + step3 = 5

Each answer builds on previous answers. That's DP — remember and reuse, never recalculate.
